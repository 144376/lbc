```python
import cv2
import os
from datetime import datetime
import numpy as np
import random
import time
# 加载人脸检测器
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# 获取摄像头
cap = cv2.VideoCapture(0, cv2.CAP_DSHOW)
cap.open(0)
faces_to_save=[]
save_folder = './python'  # 目标文件夹路径
max = 0

def save_faces(faces, save_folder):
# 创建保存人脸图像的文件夹
    if not os.path.exists(save_folder):
        os.makedirs(save_folder)

# 遍历每张人脸图像并保存
    for person_id, face in faces.items():
    # 生成唯一的文件名
        file_name = f"face_{person_id}_{datetime.now().strftime('%Y-%m-%d_%H-%M-%S')}.jpg"
        save_path = os.path.join(save_folder, file_name)
        cv2.imwrite(save_path, face)
while cap.isOpened():
# 读取帧
    ret, frame = cap.read()
    if not ret:
        cap.release()
        cv2.destroyAllWindows()
    else:
# 转换为灰度图像
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

# 检测人脸
        faces = face_cascade.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=6, minSize=(90, 90))

# 保存检测到的所有人脸图像
        if len(faces) > 0 :
            f={}
            if len(faces)>max:
                max=len(faces)
            for i, (x, y, w, h) in enumerate(faces):
# # 单独保存每张人脸
                face = frame[y:y+h, x:x+w]
                f[i]=face
            faces_to_save.append(f)
# 绘制红色矩形框
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 0, 255), 2)

# 显示帧
        cv2.imshow('Face Detection', frame)
    
# # 检测按键，如果按下esc键，退出循环
    key_pressed = cv2.waitKey(1)
    if key_pressed == 27:
        break
# 释放摄像头和关闭窗口
cap.release()
cv2.destroyAllWindows()
# 保存检测到的所有人脸图像
for faces in faces_to_save:
    if len(faces)==max:
        save_faces(faces, save_folder)
        break
```
